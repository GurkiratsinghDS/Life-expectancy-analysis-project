## Defining X and Y labels

```python
x=df.drop("Life_expectancy",axis=1)
y=df["Life_expectancy"]
```

```python
print(x.shape)
y.shape
```

## Splitting the data

```python
import random 
random.seed(20)
random.randint(1,10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=27)
# random state is like assigning one particular split,so that split will not change after each implementation
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
m1=LinearRegression()
m1.fit(x_train,y_train)
```

## to check for performance accuracy
does model perform well or not values should be near 1

## R_square

```python
print("Training data scores: ",m1.score(x_train,y_train))
print("Testing data scores: ",m1.score(x_test,y_test))
```

## evaluation metrices for error 

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
ypred_m1 = m1.predict(x_test)
```

```python
mae = mean_absolute_error(y_test,ypred_m1)
mse = mean_squared_error(y_test,ypred_m1)
rmse = np.sqrt(mean_squared_error(y_test,ypred_m1))
r2s = r2_score(y_test,ypred_m1)
```

```python
n = x_test.shape[0] # no of rows of x
p = x_test.shape[1] # no of col of x
adjusted_r2=adj_r2 = 1 - (1 - r2s) * (n - 1) / (n - p - 1)
```

```python
print("mae: ",mae)
print("mse: ",mse)
print("rmse: " , rmse)
print("r_square: ",r2s)
print("Adjusted_r2: ",adjusted_r2)
```
